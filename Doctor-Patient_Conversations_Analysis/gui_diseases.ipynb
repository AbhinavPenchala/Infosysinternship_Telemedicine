```python


# Dataset: Mapping top 15 diseases to their symptoms
disease_symptoms = {
    "Hypertension": ["Headache", "Shortness of breath", "Nosebleeds", "Dizziness"],
    "Diabetes": ["Increased thirst", "Frequent urination", "Fatigue", "Blurred vision"],
    "Cancer": ["Fatigue", "Unexplained weight loss", "Pain", "Lumps or swelling"],
    "Hepatitis": ["Jaundice", "Fatigue", "Abdominal pain", "Dark urine"],
    "Anemia": ["Fatigue", "Pale skin", "Shortness of breath", "Dizziness"],
    "Necrosis": ["Tissue death", "Skin discoloration", "Swelling", "Foul odor"],
    "HIV": ["Fever", "Weight loss", "Fatigue", "Frequent infections"],
    "Pneumonia": ["Fever", "Cough with phlegm", "Shortness of breath", "Chest pain"],
    "Tuberculosis": ["Persistent cough", "Chest pain", "Night sweats", "Weight loss"],
    "Lymphoma": ["Swollen lymph nodes", "Fever", "Night sweats", "Weight loss"],
    "Anxiety": ["Excessive worry", "Restlessness", "Fatigue", "Difficulty concentrating"],
    "Depression": ["Persistent sadness", "Loss of interest", "Sleep disturbances", "Fatigue"],
    "Obesity": ["Excess body weight", "Fatigue", "Joint pain", "Breathing difficulties"],
    "Ulcer": ["Burning stomach pain", "Nausea", "Bloating", "Heartburn"],
    "Asthma": ["Wheezing", "Shortness of breath", "Coughing", "Chest tightness"]
}

# Reverse the dictionary to map symptoms to diseases
symptom_disease_map = {}
for disease, symptoms in disease_symptoms.items():
    for symptom in symptoms:
        if symptom not in symptom_disease_map:
            symptom_disease_map[symptom] = []
        symptom_disease_map[symptom].append(disease)

# Function to find diseases based on symptoms
def find_diseases():
    input_symptoms = symptoms_entry.get().strip().split(",")  # Get input and split by comma
    input_symptoms = [symptom.strip().title() for symptom in input_symptoms]  # Normalize input
    matching_diseases = set()

    for symptom in input_symptoms:
        if symptom in symptom_disease_map:
            matching_diseases.update(symptom_disease_map[symptom])
    
    if matching_diseases:
        result_label.config(text=f"Possible Diseases:\n{', '.join(sorted(matching_diseases))}")
    else:
        messagebox.showerror("Error", "No matching diseases found for the given symptoms!")

# GUI
root = tk.Tk()
root.title("Disease Finder Based on Symptoms")
root.geometry("600x500")
root.configure(bg="#f0f8ff")  # Light blue background

# Title
title_label = tk.Label(root, text="Disease Finder", font=("Helvetica", 20, "bold"), bg="#4682b4", fg="white", padx=10, pady=10)
title_label.pack(fill="x")

# Input frame
input_frame = tk.Frame(root, bg="#f0f8ff", padx=20, pady=20)
input_frame.pack()

# Input label and entry
input_label = tk.Label(input_frame, text="Enter Symptoms (comma-separated):", font=("Arial", 12), bg="#f0f8ff")
input_label.pack(anchor="w")
symptoms_entry = tk.Entry(input_frame, width=50, font=("Arial", 12), highlightbackground="#4682b4", highlightthickness=2)
symptoms_entry.pack(pady=10)

# Search button
search_button = tk.Button(input_frame, text="Find Diseases", command=find_diseases, font=("Arial", 14, "bold"), bg="#4682b4", fg="white", relief="raised")
search_button.pack(pady=10)

# Result frame
result_frame = tk.Frame(root, bg="#f0f8ff", padx=20, pady=20)
result_frame.pack(fill="both", expand=True)

# Result label
result_label = tk.Label(result_frame, text="", font=("Arial", 12), fg="green", bg="#f0f8ff", justify="left", wraplength=500)
result_label.pack(anchor="w")

# Footer
footer_label = tk.Label(root, text="Powered by Abhinav & Gayatri", font=("Arial", 10, "italic"), bg="#4682b4", fg="white", pady=5)
footer_label.pack(fill="x", side="bottom")

# Run the GUI
root.mainloop()
```
